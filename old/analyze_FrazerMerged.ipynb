```r
PlotMeanGene <- function(gene, qs){
    
    options(repr.plot.width = 6, repr.plot.height = 3)


    plot(seq(0, (ncol(cluster.averages$RNA) - 1)), cluster.averages$RNA[gene, ], pch = 19, xlab = "cluster", ylab = paste(gene, "(mean expression)", sep = " "), las = 1,
                                                                     main = paste("Mean ", gene, " expression per cluster", sep = ""), xaxt = 'n')
    axis(side=1, at=c(0:16), cex.axis = .70)
    rect(-15, qs[,4], 17, max(cluster.averages$RNA[gene, ]+5), col = "#b2333350")
    rect(-15, qs[,3], 17, qs[,4], col = "#f2ae0f50")
    rect(-15, qs[,2], 17, qs[,3], col = "#1eb51650")
    rect(-15, -1, 17,                 qs[,2], col = "#168ccc50")


    points(seq(0, (ncol(cluster.averages$RNA) - 1)), cluster.averages$RNA[gene, ], pch = 19, xlab = "cluster", ylab = paste(gene, "(mean expression)", sep = " "), las = 1)
    
}
```

```r
PlotCTCorrelation <- function(s1, s2){


    i.s <- smoothScatter(log(cluster.averages[["RNA"]][, s1], 2), log(cluster.averages[["RNA"]][, s2], 2), pch = 16, las =1 ,
        xlab = "",     ylab = "", main = "", nrpoints = 10, ret.selection=TRUE)

    ## label the 10 very lowest-density points,the "outliers" (with obs.number):
    text(log(cluster.averages[["RNA"]][i.s[1:10],][, s1], 2), log(cluster.averages[["RNA"]][i.s[1:10],][, s2], 2), labels = rownames(cluster.averages[["RNA"]][i.s[1:10],]), cex= 0.75,
        pos = 1)
    
    
    abline(lm(cluster.averages[["RNA"]][, s2] ~ cluster.averages[["RNA"]][, s1]), col = "red")
    
    x = log(cluster.averages[["RNA"]][, s1], 2)
    x = x[! x %in% c(-Inf, Inf)]
    x2use = min(x)

    y = log(cluster.averages[["RNA"]][, s2], 2)
    y = x[! x %in% c(-Inf, Inf)]
    y2use = max(x)
    
    
    
    text(x = x2use,#min(log(cluster.averages[["RNA"]][, s1], 2)),
         y = y2use - (y2use*.1), labels = paste("r", signif(cor(cluster.averages[["RNA"]][, s2], cluster.averages[["RNA"]][, s1]), digits = 5), sep = " = "), pos = 4)
    text(x = x2use,
         y = y2use - (y2use*.2), labels = paste("p", signif(cor.test(cluster.averages[["RNA"]][, s2], cluster.averages[["RNA"]][, s1])$p.val, digits = 3), sep = " = "), pos = 4)


    
}
```
